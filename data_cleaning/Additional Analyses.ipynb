```python
post_df = pd.read_csv("datasets_annotated/post_annotated.csv")
cols = [
    "minigrid_access_life_family_impact",
"minigrid_access_life_improvement",
"minigrid_access_life_improvement_explain",
"minigrid_access_productivity_improvement",
"minigrid_benefits",
"minigrid_effect_on_household_female",
#"minigrid_problems",
] 
for col in cols:
    post_df[col] = post_df[col].astype(str)
    post_df[col].replace({"nan": "", "None": "", "none":""}, inplace=True)
post_df.head()
```

```python
pre_df = pd.read_csv("datasets_annotated/initial_annotated.csv")
pre_df.head()
```

```python
def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())

    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]

    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Join the tokens back into a string
    processed_text = ' '.join(lemmatized_tokens)

    return processed_text

# apply the function df

# initialize NLTK sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# create get_sentiment function

def get_sentiment(text, threshold=0.5):
    scores = analyzer.polarity_scores(text)
    sentiment = 1 if scores['pos'] > threshold else 0

    return sentiment


# apply get_sentiment function
#grp = (post_df['country'] != post_df['country'].shift()).cumsum().rename('order')
col = "minigrid_effect_on_household_female"

kenya = pd.DataFrame(post_df.groupby(['kenya_community'])[col].apply(' | '.join).sort_index(level=1)).reset_index()
kenya.rename({"kenya_community": "community"}, inplace=True, axis=1)
nigeria = pd.DataFrame(post_df.groupby(['nigeria_community'])[col].apply(' | '.join).sort_index(level=1)).reset_index()
nigeria.rename({"nigeria_community": "community"}, inplace=True, axis=1)
df = pd.concat([kenya, nigeria], axis=0, ignore_index = True)
#df[col]
#col = "minigrid_benefits"
df[col+"_pre_process"] = df[col].astype(str).apply(preprocess_text)
df['sentiment'] = df[col+"_pre_process"].apply(get_sentiment)
df
```
